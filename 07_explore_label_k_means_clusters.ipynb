## Functions

```python
def print_features(cluster_n, n_features):
    features = centroids[cluster_n][:n_features]
    for feature in features:
        print(feature_names[feature])

def print_df(cluster_n, df=df, sample_size=30):
    mask = df['new_km_150'] == cluster_n
    return df[mask][['title', 'new_km_150']].sample(sample_size)

def add_labels(val):
    return cluster_labels[val]
```
